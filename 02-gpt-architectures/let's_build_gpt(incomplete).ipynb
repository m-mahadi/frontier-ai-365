```python
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(''.join(chars))
print(vocab_size)
```

```python
stoi = {ch:i for i, ch in enumerate(chars)}
itos = {i:ch for i, ch in enumerate(chars)}
encode = lambda s: [stoi[c] for c in s]
decode = lambda l: ''.join([itos[i] for i in l])

print(encode("hii there"))
print(decode(encode("hii there")))
```

```python
#let's now encode the entire text dataset tand store into a torch.tensor
import torch
data = torch.tensor(encode(text), dtype=torch.long)
print(data.shape, data.dtype)
print(data[:1000])
```

```python
#let's now split up the data into train and validation setss
n = int(0.9*(len(data)))
train_data = data[:n]
val_data = data[n:]
```

```python
block_size = 8
train_data[:block_size+1]
```

```python
x = train_data[:block_size]
y = train_data[1:block_size+1]
for t in range(block_size):
  context = x[:t+1]
  target = y[t]
  print(f"when input is {context} the target: {target  }")
```

```python
torch.manual_seed(1337)
batch_size = 4 #how many independent sequenced will we process in parallel?
block_size = 8 #what is the maximum context length

def get_batch(split):
  #generate a small batch of data of inputs x and targets y
  data = train_data if split == 'train' else val_data
  ix = torch.randint(len(data) - block_size, (batch_size,))
  x = torch.stack([data[i:i+block_size] for i in ix])
  y = torch.stack([data[i+1:i+block_size+1] for i in ix])
  return x, y

xb, yb = get_batch('train')
print('inputs:')
print(xb.shape)
print(xb)
print('targets:')
print(yb.shape)
print(yb)

print('-----')

for b in range(batch_size):
  for t in range(block_size):
    context = xb[b, :t+1]
    target = yb[b, t]
    print(f"when input is {context.tolist()} the target: {target}")
```

```python
import torch
import torch.nn as nn
from torch.nn import functional as F
torch.manual_seed(1337)

class BigramLanguageModel(nn.Module):
  def __init__(self, vocab_size):
    super().__init__()
    #each token directly reads off the logits for the next token from the lookup tabke
    self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)

  def forward(self, idx,  targets=None):
    #idx and targets are both (B, T) tensor of integers
    logits = self.token_embedding_table(idx) #(B, T, C)

    if targets is None:
      loss = None
    else:
      B, T, C = logits.shape
      logits = logits.view(B*T, C)
      targets = targets.view(-1)

      loss = F.cross_entropy(logits, targets)

    return logits, loss

  def generate(self, idx, max_new_tokens):
    #idx is (B, T) array of indicess in the current context
    for _ in range(max_new_tokens):
      #get the pred
      logits, loss = self(idx)
      #focus only on the last step
      logits = logits[:, -1, :] #becomes (B,C)
      #apply softmax to probablities
      probs = F.softmax(logits, dim=-1) #(B, C)
      #sample from the distribution
      idx_next = torch.multinomial(probs, num_samples=1) #(B, 1)
      #append sampled index to the running sequence
      idx = torch.cat((idx, idx_next), dim=1)  #(B, T+1)

    return idx

m = BigramLanguageModel(vocab_size)
logits, loss = m(xb, yb)
print(logits.shape)
print(loss)
idx = torch.zeros((1,1), dtype=torch.long)
print(decode(m.generate(idx, max_new_tokens=100)[0].tolist()))
```

```python
#create a pytorch optimizer
optimizer = torch.optim.AdamW(m.parameters(), lr=1e-1)
```

```python
batch_size = 32
for steps in range(10000):
  #sample a data of batch
  xb, yb = get_batch('train')

  #evaluate the loss
  logits, loss = m(xb, yb)
  optimizer.zero_grad(set_to_none=True)
  loss.backward()
  optimizer.step()

print(loss.item())
```
